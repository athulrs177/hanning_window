```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt

import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.colors
import matplotlib

from apply_hanning_window import apply_hann_window_lon
```

```python
fili = 'tropics_imerg_v07_06UTC_daily_accum_1.0deg_2000_2023.nc'

f = xr.open_dataset(fili)
f
```

```python
data = f.precipitation.transpose('time','lat','lon')
```

```python
center_lon = 20
window_size = 121

LON = f'{center_lon}E'
if center_lon<0:
    LON = f'{center_lon}W'

windowed_data = apply_hann_window_lon(da=data, center_lon=center_lon, window_size=window_size)
```

```python
windowed_data
```

```python
fig, axs = plt.subplots(nrows=2, ncols=1, subplot_kw=dict(projection=ccrs.PlateCarree()), sharex=True, sharey=True)

x_ticks = [-50,-30,-10,10,30,50,70,90]
y_ticks = [-30,-10,10,30]
ext = [-50,90,-30,30]

fontsize=10

ax1, ax2 = axs.flatten()

cmap = 'Blues'
levels = np.arange(0,12.5,0.5)

lats = data.sel(time='02-05-2023',lon=slice(-50, 90)).lat.to_index()
lons = data.sel(time='02-05-2023',lon=slice(-50, 90)).lon.to_index()

cont1 = ax1.contourf(lons, lats, data.sel(time='02-05-2023',lon=slice(-50, 90)).squeeze(),
              levels=levels, cmap=cmap, extend='max', transform=ccrs.PlateCarree(),)
cont1 = ax2.contourf(lons, lats, windowed_data.sel(time='02-05-2023',lon=slice(-50, 90)).squeeze(),
              levels=levels, cmap=cmap, extend='max', transform=ccrs.PlateCarree(),)


ax1.coastlines()
ax1.add_feature(cfeature.BORDERS.with_scale('50m'))
ax1.set_extent(ext,crs=ccrs.PlateCarree())
yt1 = ax1.set_yticks(y_ticks, crs=ccrs.PlateCarree())
# xt1 = ax1.set_xticks(x_ticks, crs=ccrs.PlateCarree())
ax1.set_yticklabels(y_ticks, rotation=0, fontsize=fontsize)
# ax1.set_xticklabels(x_ticks, rotation=0, fontsize=fontsize)
ax1.tick_params(axis='both', which='major', labelsize=fontsize)
ax1.set_title('a) No filtering')

ax2.coastlines()
ax2.add_feature(cfeature.BORDERS.with_scale('50m'))
ax2.set_extent(ext,crs=ccrs.PlateCarree())
yt2 = ax2.set_yticks(y_ticks, crs=ccrs.PlateCarree())
xt2 = ax2.set_xticks(x_ticks, crs=ccrs.PlateCarree())
ax2.set_yticklabels(y_ticks, rotation=0, fontsize=fontsize)
ax2.set_xticklabels(x_ticks, rotation=0, fontsize=fontsize)
ax2.tick_params(axis='both', which='major', labelsize=fontsize)
ax2.set_title(f'b) Hanning window applied at {center_lon}E')


cax2 = fig.add_axes([0.850, 0.18, 0.02, 0.645])
fig_clb1=fig.colorbar(cont1, ax=cax2, orientation='vertical',shrink=0.01,cax=cax2)
fig_clb1.set_label(r'$mmday^{-1}$',size=fontsize)
fig_clb1.ax.tick_params(labelsize=fontsize)

plt.show()
```
